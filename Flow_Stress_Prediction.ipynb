```python
strain_rate = [0.001,0.01,0.1,1]
strain = np.arange(0, 1.2, 0.005)
numpy_arr = np.empty((0, 6), float)
temperature = [1253]
for temp in temperature:
    for i in strain_rate:
        for j in strain:
            numpy_arr = np.append(numpy_arr, np.array([[temp,i,j,math.log(i),math.exp(2000/(8.314*temp)),i*math.exp(2000/(8.314*temp))]]), axis=0)
df6 = pd.DataFrame(numpy_arr,columns=['Temperature', 'Strain_rate','True_Strain','log(Strain_rate)','exp(Q/RT)','Z'])
df6
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# #comparison plot between experiment and predicted data
fig, axes = plt.subplots(nrows=2,ncols=2, figsize=(12, 12), tight_layout=True)
sns.scatterplot(x = 'True_Strain',y = 'True_Stress',data = data_1,hue = 'Temperature',palette = 'dark',ax = axes[0][0]).set(title="strain rate = 0.001")
sns.scatterplot(x = 'True_Strain',y = 'True_Stress',data = data_2,hue = 'Temperature',palette = 'dark',ax = axes[0][1]).set(title="strain rate = 0.01")
sns.scatterplot(x = 'True_Strain',y = 'True_Stress',data = data_3,hue = 'Temperature',palette = 'dark',ax =axes[1][0]).set(title="strain rate = 0.1")
sns.scatterplot(x = 'True_Strain',y = 'True_Stress',data = data_4,hue = 'Temperature',palette = 'dark',ax = axes[1][1]).set(title="strain rate = 1")
plt.legend(loc='upper right')
```
